```python
import torch;
import pandas as pd;
#supress pandas warnings
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# read the dataset from the file updated_locational_prices.csv
data = pd.read_csv('updated_locational_prices.csv');
data
```

```python
# get the time, start_location, end_location, price, and duration columns
needed = data[['time', 'start_location', 'end_location', 'total_price', 'duration']];
needed

# strip the s from the duration column and put it in the column name 
needed['duration'] = needed['duration'].str.strip('s');
needed.rename(columns = {'duration': 'duration_seconds'}, inplace = True);
# convert the duration column to a float
needed['duration_seconds'] = needed['duration_seconds'].astype(float);
needed
```

```python
# Get the average duration and total_price for each start location
# group the data by the start location
grouped = needed.groupby('start_location');
# get the average duration and total_price for each start location
average_duration = grouped['duration_seconds'].mean();
average_price = grouped['total_price'].mean();
# create a new dataframe to store the average duration and total_price
average_df = pd.DataFrame({'average_duration': average_duration, 'average_price': average_price});
average_df
```
